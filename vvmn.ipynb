```python
import os
import pandas as pd

datasets_path = '/content/drive/MyDrive/VVMN/datasets/'

# List of country codes
country_codes = ['US', 'RU', 'MX', 'KR', 'JP', 'IN', 'GB', 'FR', 'DE', 'CA', 'BR']

# Create an empty list to store DataFrames for each country
dfs = []

# Load and append datasets for each country
for country_code in country_codes:
    file_path = os.path.join(datasets_path, f'{country_code}_youtube_trending_data.csv')
    country_df = pd.read_csv(file_path)

    # Add a new column indicating the country code
    country_df['country_code'] = country_code

    # Append to the list of DataFrames
    dfs.append(country_df)

# Concatenate the list of DataFrames into one
combined_df = pd.concat(dfs, ignore_index=True)

# List of columns to keep
relevant_columns = ['country_code', 'view_count', 'likes', 'dislikes', 'comment_count', 'channelTitle', 'categoryId', 'title','publishedAt','trending_date',]

# Keep only relevant columns
combined_df = combined_df[relevant_columns]

# Save the combined dataset to a new CSV file
combined_df.to_csv('/content/drive/MyDrive/VVMN/combined_youtube_data.csv', index=False)
```

```python
# Convert 'view_count' to numeric (ignoring errors to handle mixed types)
combined_df['view_count'] = pd.to_numeric(combined_df['view_count'], errors='coerce')

# Sort the combined_df by views in descending order
sorted_df = combined_df.sort_values('view_count', ascending=False)

# Top 50 videos from the entire dataset
top_50_df = sorted_df.head(50)

top_50_df
```

```python
# Set display format for floating-point numbers
pd.options.display.float_format = '{:,.0f}'.format

# Group by country and categoryId, summing up the metrics for each genre
genre_preferences = combined_df.groupby(['country_code', 'categoryId'])[['view_count', 'likes', 'dislikes', 'comment_count']].sum().reset_index()

# Sort by view_count in descending order to identify the most popular genres
genre_preferences = genre_preferences.sort_values(by='view_count', ascending=False)

# Display the top genres for each country
for country_code in genre_preferences['country_code'].unique():
    top_genres_country = genre_preferences[genre_preferences['country_code'] == country_code].head(5)
    print(f"Top genres in {country_code}:")
    print(top_genres_country)
    print("\n")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# Convert 'trending_date' and 'publishedAt' to datetime objects
combined_df['trending_date'] = pd.to_datetime(combined_df['trending_date'])
combined_df['publishedAt'] = pd.to_datetime(combined_df['publishedAt'])

# Calculate the time it takes for videos to become trending
combined_df['time_to_trend'] = (combined_df['trending_date'] - combined_df['publishedAt']).dt.days

# Sample data
sampled_data = combined_df.sample(n=50)

# Map categoryId to genre names
category_mapping = {
    1: 'Film & Animation', 2: 'Autos & Vehicles', 10: 'Music', 15: 'Pets & Animals',
    17: 'Sports', 18: 'Short Movies', 19: 'Travel & Events', 20: 'Gaming',
    21: 'Videoblogging', 22: 'People & Blogs', 23: 'Comedy', 24: 'Entertainment',
    25: 'News & Politics', 26: 'Howto & Style', 27: 'Education', 28: 'Science & Technology',
    29: 'Nonprofits & Activism', 30: 'Movies', 31: 'Anime/Animation', 32: 'Action/Adventure',
    33: 'Classics', 34: 'Comedy', 35: 'Documentary', 36: 'Drama', 37: 'Family',
    38: 'Foreign', 39: 'Horror', 40: 'Sci-Fi/Fantasy', 41: 'Thriller', 42: 'Shorts',
    43: 'Shows', 44: 'Trailers'
}

# Create a directed graph
G = nx.DiGraph()

# Add nodes for genres and countries
genres = set(sampled_data['categoryId'].map(category_mapping))
countries = set(sampled_data['country_code'])

G.add_nodes_from(genres, bipartite=0)  # Genres
G.add_nodes_from(countries, bipartite=1)  # Countries

# Connect genres to countries with directed edges based on the average time to trend
for genre in genres:
    for country in countries:
        avg_time = sampled_data[(sampled_data['categoryId'].map(category_mapping) == genre) & (sampled_data['country_code'] == country)]['time_to_trend'].mean()
        if not pd.isnull(avg_time):
            G.add_edge(genre, country, weight=avg_time)

pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, font_size=8, font_color="black", node_size=400, node_color='skyblue', font_weight='bold', edge_color='black', width=0.5)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Map category IDs to genre names
sampled_data['genre'] = sampled_data['categoryId'].map(category_mapping)

# Create a pivot table to get average time for each genre in each country
pivot_table = sampled_data.pivot_table(values='time_to_trend', index='genre', columns='country_code', aggfunc='mean')

# Set up the heatmap using Seaborn
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='coolwarm', annot=True, fmt=".1f", linewidths=.5)

# Set plot labels and title
plt.xlabel('Country Code')
plt.ylabel('Video Genre')
plt.title('Average Time to Trend by Genre and Country')

# Show the plot
plt.show()
```
